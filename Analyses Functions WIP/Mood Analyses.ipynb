```python
logcortlist = []
meanvaslist = [vasalertcortmeanlist, vassadcortmeanlist, vastensecortmeanlist,
                vaseffortcortmeanlist,
              vashappycortmeanlist, vaswearycortmeanlist, vascalmcortmeanlist,
              vassleepycortmeanlist, vasglobalvigorcortmeanlist, vasglobalaffectcortmeanlist]
moodlist = []
deletelist = []
diftrierlist = []
agelist = []; genderlist = []
for i, x in enumerate(cortsubslist):
    logcortlist.append([x, []])
    moodlist.append([x])
    agelist.append([x])
    genderlist.append([x])
    try:
        moodlist[i].append(cortdf.loc[cortdf['Record ID'] == int(x), 'Mood Condition'].iloc[0])
        agelist[i].append(cortdf.loc[cortdf['Record ID'] == int(x), 'Age'].iloc[0])
        genderlist[i].append(cortdf.loc[cortdf['Record ID'] == int(x), 'Gender'].iloc[0])
        for y in range(1,12):
            logcortlist[i][1].append(cortdf.loc[(cortdf['Record ID'] == int(x)) & (cortdf['General Time'] == y), 'Log Mean (µg/dL)'].iloc[0])
    except:
        deletelist.append(i)
loop=0
for x in deletelist:
    logcortlist.pop(x-loop)
    genderlist.pop(x-loop)
    moodlist.pop(x-loop)
    for i, y in enumerate(meanvaslist):
        meanvaslist[i].pop(x-loop)
    vashappycortstdlist.pop(x-loop)
    loop +=1
diftrierlist = [i-j for i,j in zip([x[1][8] for x in logcortlist],[x[1][7] for x in logcortlist])]
diftrierlistz = list(stats.mstats.zscore(diftrierlist))

deletelist = []
for i, x in enumerate(diftrierlistz):
    if x > 3 or x < -3:
        deletelist.append(i)      
        print(x)
loop = 0
for x in deletelist:
    moodlist.pop(x-loop)
    genderlist.pop(x-loop)
    diftrierlistz.pop(x-loop)
    for i, y in enumerate(meanvaslist):
        meanvaslist[i].pop(x-loop)
    vashappycortstdlist.pop(x-loop)
    loop +=1
```

```python
#Edit These
SAVE = False
var1 = diftrierlistz
var2 = [y[1] for y in vashappycortstdlist]
var1label = "Log Cortisol Delta (z-score)"
var2label = "Average VAS Global Affect Score"
title = "Log Cortisol Delta vs.\nAverage VAS Global Affect Score (N=%s)" % len(var1)
savename = "Analyses/VAS/VASGlobalAffect-Stress.png"

#--------------------------------------------
table = pd.DataFrame(
    {var1label: var1,
    var2label: var2
    })
for x in table.columns:
    table = table[np.isfinite(table[x])]

lm = seaborn.lmplot(x=var1label, y=var2label, palette=('r'), data=table, legend_out=False)
ax = mp.pyplot.gca()
ax.set_title("%s\nr=%.4f, p=%.4f" % (title,pearsonr(table[var1label],table[var2label])[0],pearsonr(table[var1label],table[var2label])[1]))
print("r = %s, p = %s" % pearsonr(table[var1label],table[var2label]))
if SAVE == True:
    lm.savefig(root / savename, bbox_inches='tight')
```

```python
#Edit These
SAVE = True
var1 = diftrierlistz
var2 = [y[1] for y in meanvaslist[9]]
groupvar = [y[1] for y in genderlist]
MEANSPLIT = False
var1label = "Pre-Post TSST Difference in Cortisol (µg/dL)"
var2label = "Average VAS Global Affect Score"
highgrouplabel = "High EQI Total Score"
lowgrouplabel = "Low EQI Total Score"
title = "Log Cortisol Delta vs.\nAverage VAS Global Affect Score"
savename = "Analyses/VAS/Gender/VASGlobalAffect-Stress-GenderSplit.png"

#--------------------------------------------
table = pd.DataFrame(
    {var1label: var1,
    var2label: var2,
    'z_raw': groupvar,
    'z_group': groupvar
    })

grouplist = []

for i, x in enumerate(table.z_raw):
    if MEANSPLIT == True:
        if x > statistics.mean(groupvar):
            grouplist.append(highgrouplabel)
        else:
            grouplist.append(lowgrouplabel)
    else:
        grouplist.append(groupvar[i])

for x in list(unique_everseen(grouplist)):
    grouplist = [w.replace(x, x + ' (N=%s)' % grouplist.count(x)) for w in grouplist]
table['z_group'] = grouplist

seaborn.set(rc={'figure.figsize':(300,300)})
#seaborn.reset_orig()

lm = seaborn.lmplot(x=var1label, y=var2label, hue = 'z_group', data=table, legend=False)
ax = mp.pyplot.gca()
ax.set_title(title)
mp.pyplot.legend(bbox_to_anchor=(1, 1), loc=2)

for x in list(unique_everseen(grouplist)):
    print("%s" % x + " Group: r = %s, p = %s" % (pearsonr(table[var1label][table['z_group'] == x],table[var2label][table['z_group'] == x])))
if MEANSPLIT == True:
    print("Mean of Grouping Variable: %.4f" % statistics.mean(groupvar))
if SAVE == True:
    lm.savefig(root / savename, bbox_inches='tight')
```
